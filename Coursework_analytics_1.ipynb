```python
import pandas as pd
import numpy as np

# Загрузка данных из файлов CSV в DataFrame
courses_df = pd.read_csv('courses.csv')
students_df = pd.read_csv('students.csv')
course_contents_df = pd.read_csv('course_contents.csv')
progresses_df = pd.read_csv('progresses.csv')
progress_phases_df = pd.read_csv('progress_phases.csv')

# Выберем случайные 5% строк из каждого DataFrame
courses_df = courses_df.sample(frac=0.50, random_state=42)
students_df = students_df.sample(frac=0.50, random_state=42)
course_contents_df = course_contents_df.sample(frac=0.50, random_state=42)
progresses_df = progresses_df.sample(frac=0.50, random_state=42)
progress_phases_df = progress_phases_df.sample(frac=0.50, random_state=42)
```

```python
merged_courses = pd.merge(courses_df, course_contents_df, left_on='id', right_on='course_id')
merged_progresses = pd.merge(progresses_df, progress_phases_df, left_on='id', right_on='progress_id')
merged_data = pd.merge(merged_progresses, merged_courses, on='course_id')
merged_data = merged_data.drop(['id_x', 'module_number_x', 'lesson_number_x', 'Unnamed: 0', 'id_y'], axis=1)
merged_data = merged_data.rename(columns={'module_number_y': 'module_number', 'lesson_number_y': 'lesson_number'})
merged_data.to_csv('merged_data.csv', index=False)
```

```python
merged_data = pd.read_csv('merged_data.csv')
```

```python
merged_data_with_students = pd.merge(merged_data, students_df, left_on='student_id', right_on='id')
merged_data_with_students.drop('id', axis=1, inplace=True)
merged_data_with_students.to_csv('merged_data_with_students.csv', index=False)
```

```python
merged_data_with_students = pd.read_csv("merged_data_with_students.csv")
```

```python
# Общее количество курсов в датасете
total_courses = merged_data_with_students['course_id'].nunique()
print(f"Общее количество курсов: {total_courses}")

# Количество модулей на каждом курсе
modules_per_course = merged_data_with_students.groupby('course_id')['module_number'].nunique()
print("Количество модулей на каждом курсе:")
print(modules_per_course)

# Количество уроков в каждом модуле на каждом курсе
lessons_per_module_per_course = merged_data_with_students.groupby(['course_id', 'module_number'])['lesson_number'].nunique()
print("Количество уроков в каждом модуле на каждом курсе:")
print(lessons_per_module_per_course)

# Медианное количество уроков в модуле на каждом курсе
median_lessons_per_module_per_course = merged_data_with_students.groupby(['course_id', 'module_number'])['lesson_number'].median()
print("Медианное количество уроков в модуле на каждом курсе:")
print(median_lessons_per_module_per_course)

# Количество учеников на каждом курсе
students_per_course = merged_data_with_students.groupby('course_id')['student_id'].nunique()
print("Количество учеников на каждом курсе:")
print(students_per_course)

# Фильтрация и подсчет возраста студентов
# Предположим, что возраст студентов представлен в годах и в столбце 'birthday'
# Установим диапазон для правдоподобного возраста (например, 18-100 лет)
valid_age_range = (18, 100)
merged_data_with_students['birthday'] = pd.to_datetime(merged_data_with_students['birthday'], errors='coerce')
merged_data_with_students = merged_data_with_students.dropna(subset=['birthday'])
merged_data_with_students['age'] = (pd.to_datetime('now') - merged_data_with_students['birthday']).astype('<m8[Y]')
filtered_students = merged_data_with_students[(merged_data_with_students['age'] >= valid_age_range[0]) & (merged_data_with_students['age'] <= valid_age_range[1])]

# Статистика возраста студентов
min_age = filtered_students['age'].min()
max_age = filtered_students['age'].max()
mean_age = filtered_students['age'].mean()
median_age = filtered_students['age'].median()

print(f"Минимальный возраст студентов: {min_age} лет")
print(f"Максимальный возраст студентов: {max_age} лет")
print(f"Средний возраст студентов: {mean_age:.2f} лет")
print(f"Медианный возраст студентов: {median_age} лет")

# Минимальный, максимальный, средний, медианный возраст студентов на каждом курсе
age_stats_per_course = filtered_students.groupby('course_id')['age'].agg(['min', 'max', 'mean', 'median'])
print("Статистика возраста студентов на каждом курсе:")
print(age_stats_per_course)
```

```python
import matplotlib.pyplot as plt
# Группировка данных и подсчет уникальных студентов на каждом курсе
students_per_course = merged_data_with_students.groupby('course_id')['student_id'].nunique()

# Создание bar-графика
plt.figure(figsize=(12, 6))  # Установка размеров графика

# Построение bar-графика
students_per_course.plot(kind='bar')

# Настройка меток оси X для лучшей читаемости
plt.xticks(rotation=45, ha='right')

# Настройка заголовка и меток осей
plt.title('Количество студентов на каждом курсе')
plt.xlabel('Курс')
plt.ylabel('Количество студентов')

# Отображение графика
plt.tight_layout()
plt.show()
```

```python
# Группировка данных по курсам и сферам, подсчет количества студентов
grouped_data = merged_data_with_students.groupby(['title', 'field'])['student_id'].nunique().unstack()

# Сортировка данных по количеству студентов на каждом курсе
sorted_data = grouped_data.sum(axis=1).sort_values(ascending=False)

# Подготовка таблицы данных для графика через unstack()
plot_data = grouped_data.loc[sorted_data.index]

# Создание горизонтального bar-графика
colors = ['b', 'g', 'r', 'c']  # Цвета для каждой сферы
plot_data.plot(kind='barh', stacked=True, color=colors, alpha=0.5, figsize=(10, 6))

# Добавление линии медианы
median_line = plt.axvline(x=plot_data.median().mean(), color='purple', linestyle='--', linewidth=2, label='Медиана')

# Настройка легенды
plt.legend(handles=[median_line], loc='upper right')

# Настройка заголовка и меток осей
plt.title('Количество студентов на каждом курсе')
plt.xlabel('Количество студентов')
plt.ylabel('Курсы')

# Отображение графика
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузим данные из файла с предоставленными названиями столбцов
data = pd.read_csv('merged_data_with_students.csv')

# Функция для определения месяца из даты
def get_month(date_str):
    return pd.to_datetime(date_str).strftime('%Y-%m')

# Создаем датасет с «адресами» первых домашних работ на курсах
first_homeworks = data[data['is_homework'] == True].groupby(['course_id', 'module_number'])['start_date'].min().reset_index()
first_homeworks['start_month'] = first_homeworks['start_date'].apply(get_month)

# Фильтруем датасет только по первым домашним работам в каждом месяце
first_homeworks = first_homeworks.groupby(['course_id', 'start_month'])['start_date'].count().reset_index()
first_homeworks.rename(columns={'start_date': 'students_count'}, inplace=True)

# Создаем сводную таблицу с количеством студентов в каждом месяце для каждого курса
pivot_table = first_homeworks.pivot(index='start_month', columns='course_id', values='students_count')

# Заполняем пропущенные значения нулями (если в каком-то месяце не было первых домашних работ на курсе)
pivot_table.fillna(0, inplace=True)

# Построим line-graph с приростом студентов в каждом месяце для каждого курса (15 графиков)
for course_id in pivot_table.columns:
    plt.figure(figsize=(10, 6))
    plt.plot(pivot_table.index, pivot_table[course_id], label=f'Курс {course_id}')
    plt.title(f'Прирост студентов на курсе {course_id}')
    plt.xlabel('Месяц')
    plt.ylabel('Количество студентов')
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Построим line-graph с несколькими линиями, отражающими прирост студентов в каждом месяце для каждого курса (15 линий на графике)
plt.figure(figsize=(10, 6))
for course_id in pivot_table.columns:
    plt.plot(pivot_table.index, pivot_table[course_id], label=f'Курс {course_id}')
plt.title('Прирост студентов в каждом месяце для каждого курса')
plt.xlabel('Месяц')
plt.ylabel('Количество студентов')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных
data = pd.read_csv('merged_data_with_students.csv')

# Преобразование столбцов с датами в формат datetime
data['start_date'] = pd.to_datetime(data['start_date'])
data['finish_date'] = pd.to_datetime(data['finish_date'])

# Удаление записей без финишной даты
data.dropna(subset=['finish_date'], inplace=True)

# Рассчет времени выполнения каждой домашней работы
data['time_to_complete'] = data['finish_date'] - data['start_date']

# Группировка по курсам и модулям
module_times = data.groupby(['course_id', 'module_number'])['time_to_complete'].agg(['min', 'max', 'mean', 'median'])

# Задание 1: Рассчет минимального, максимального, среднего, медианного времени прохождения каждого модуля
print("Задание 1 - Рассчет минимального, максимального, среднего, медианного времени прохождения каждого модуля")
print(module_times)

# Задание 2: Построение графика с медианным временем прохождения каждого модуля для каждого курса
print("Задание 2 - Построение графика с медианным временем прохождения каждого модуля для каждого курса")
for course_id in module_times.index.levels[0]:
    module_times.loc[course_id]['median'].plot(label=f'Курс {course_id}', figsize=(10, 6))
    plt.title(f'Медианное время прохождения модулей по курсам')
    plt.xlabel('Модуль')
    plt.ylabel('Время прохождения (дни)')
    plt.legend()
    plt.tight_layout()
    plt.show()

# Задание 3: Рассчет медианного времени выполнения домашней работы по месяцам для каждого курса
print("Задание 3 - Рассчет медианного времени выполнения домашней работы по месяцам для каждого курса")
data['start_month'] = data['start_date'].dt.strftime('%Y-%m')
median_times_by_month = data.groupby(['course_id', 'start_month'])['time_to_complete'].median()
print('Рассчет медианного времени выполнения домашней работы по месяцам для каждого курса: ', median_times_by_month)

# Задание 4: Построение графика с медианным временем выполнения домашней работы по месяцам для каждого курса
print("Задание 4 - Построение графика с медианным временем выполнения домашней работы по месяцам для каждого курса")
plt.figure(figsize=(10, 6))
for course_id in median_times_by_month.index.levels[0]:
    median_times_by_month[course_id].plot(label=f'Курс {course_id}')

plt.title('Медианное время выполнения домашней работы по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Время выполнения (дни)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
data = pd.read_csv('merged_data_with_students.csv')

# Рассчитаем конверсию перехода студентов из одного модуля в другой на каждом курсе
module_conversion = data.groupby(['course_id', 'module_number'])['student_id'].agg('nunique').reset_index()
module_conversion['conversion'] = module_conversion['student_id'] / module_conversion['student_id'].shift(1)
module_conversion['conversion'] = module_conversion['conversion'].fillna(0)

# Выведем таблицу с конверсией перехода для каждого модуля на каждом курсе
print(module_conversion)

# Построим bar-chart с отражением конверсии перехода студентов из одного модуля в другой для каждого курса (15 графиков)
for course_id in module_conversion['course_id'].unique():
    course_data = module_conversion[module_conversion['course_id'] == course_id]
    plt.figure(figsize=(10, 6))
    plt.bar(course_data['module_number'], course_data['conversion'], color='skyblue')
    plt.title(f'Конверсия перехода студентов на курсе {course_id}')
    plt.xlabel('Номер модуля')
    plt.ylabel('Конверсия')
    plt.xticks(course_data['module_number'])
    plt.tight_layout()
    plt.show()
```

```python
# Построим горизонтальный bar-chart с отражением конверсии перехода студентов из одного модуля в другой на каждом курсе (15 графиков)
for course_id in module_conversion['course_id'].unique():
    course_data = module_conversion[module_conversion['course_id'] == course_id]
    plt.figure(figsize=(10, 6))
    plt.barh(course_data['module_number'], course_data['conversion'], color='blue', alpha=0.1)
    plt.title(f'Конверсия перехода студентов на курсе {course_id}')
    plt.xlabel('Конверсия')
    plt.ylabel('Номер модуля')
    plt.yticks(course_data['module_number'])
    plt.axvline(course_data['conversion'].median(), color='red', linestyle='dashed', label='Медиана')
    plt.legend()
    plt.tight_layout()
    plt.show()
```
